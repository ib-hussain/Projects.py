# Exploratory Data Analysis (EDA)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(style="whitegrid")
```

## 1) Load dataset

```python
df = sns.load_dataset('titanic')
print("Shape:", df.shape)
df.head()
```

## 2) Inspect dataset

```python
df.info()
print("\nMissing values per column:\n")
print(df.isna().sum())
df.describe(include='all').T
```

## 3) Data Cleaning

```python
# Drop high-null columns (deck has too many NaN)
df = df.drop(columns=['deck'])

# Fill missing embark_town with mode
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])

# Fill age with median
df['age'] = df['age'].fillna(df['age'].median())

# Drop rows with null 'embarked' (only 2 rows)
df = df.dropna(subset=['embarked'])

# Convert categorical columns to category dtype
cat_cols = df.select_dtypes(include=['object']).columns
for c in cat_cols:
    df[c] = df[c].astype('category')

print("Cleaned shape:", df.shape)
df.head()
```

## 4) Exploratory Data Analysis

```python
# Distribution of age
plt.figure(figsize=(7,4))
sns.histplot(df['age'], bins=30, kde=True, color='steelblue')
plt.title('Age Distribution')
plt.show()

# Survival by sex
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sex', hue='survived')
plt.title('Survival by Sex')
plt.show()

# Survival by passenger class
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='class', hue='survived')
plt.title('Survival by Passenger Class')
plt.show()

# Correlation heatmap of numeric variables
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 5) Key Findings

- **Females survived at much higher rates** than males.
- **Higher passenger class (1st class)** had higher survival compared to 3rd class.
- Age distribution shows most passengers were between 2040 years old.
- Fare is strongly correlated with passenger class.
